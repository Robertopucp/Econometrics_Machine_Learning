## Carga de las imágenes: Fruit Labelling

```
import numpy as np 
import pandas as pd
import os
import cv2
import matplotlib.pyplot as plt
from tqdm import tqdm
from random import shuffle
from keras.utils  import to_categorical
import pickle
import matplotlib.pyplot as plt
import matplotlib as mpl
```

### Cargamos un subset pequeño para visualizar las imágenes

```
def load_rand():
    X=[]
    dir_path='dataset/train'
    list_names = []
    for sub_dir in tqdm(os.listdir(dir_path)):
        print("Analyzing this subdir:",sub_dir)
        path_main=os.path.join(dir_path,sub_dir)
        i=0
        list_names.append(path_main[14:])
        for img_name in os.listdir(path_main):
            if i>=6:
                break
            img=cv2.imread(os.path.join(path_main,img_name))
            img=cv2.resize(img,(100,100))
            img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
            X.append(img)
            i+=1
    return X, list_names
```

```
x_aleat, list_names = load_rand()
x_aleat = np.array(x_aleat)
x_aleat.shape
```

```
def show_subpot(X,list_fruits,title=False,Y=None):
    if X.shape[0]==36:
        f, ax= plt.subplots(6,6, figsize=(20,30))
        for i,img in enumerate(X):
            ax[i//6][i%6].imshow(img, aspect='auto')
            if title==False:
                ax[i//6][i%6].set_title(list_fruits[i//6])
            elif title and Y is not None:
                ax[i//6][i%6].set_title(Y[i])
        plt.show()
    else:
        print('No puedo plotear. La variable ingresada debe tener 36 imágenes.')
```

```
show_subpot(x_aleat, list_names)
```

### Cargamos el `train` dataset

```
def load_fruit_labels(subdir):
    assert ((subdir == "train")| (subdir == "test")), "Sólo se pueden ingresar subdirectorios que sean `train` o `test`."
    quality=['apple', 'banana','orange']
    X,Y=[],[]
    z=[]
    baseDirectory="dataset/"
    for cata in tqdm(os.listdir(baseDirectory+subdir+"/")):
        if quality[0] in cata:
            path_main=os.path.join(baseDirectory+subdir+"/",cata)
            for img_name in os.listdir(path_main):
                img=cv2.imread(os.path.join(path_main,img_name))
                img=cv2.resize(img,(100,100))
                img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
                z.append([img,0])
        elif quality[1] in cata:
            path_main=os.path.join(baseDirectory+subdir+"/",cata)
            for img_name in os.listdir(path_main):
                img=cv2.imread(os.path.join(path_main,img_name))
                img=cv2.resize(img,(100,100))
                img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
                z.append([img,1])
        elif quality[2] in cata:
            path_main=os.path.join(baseDirectory+subdir+"/",cata)
            for img_name in os.listdir(path_main):
                img=cv2.imread(os.path.join(path_main,img_name))
                img=cv2.resize(img,(100,100))
                img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
                z.append([img,2])
        
    print('Shuffling your data.....')
    shuffle(z)
    for images, labels in tqdm(z):
        X.append(images);Y.append(labels)
    return X,Y 
```

```
x_fruits_train_color,y_fruits_train=load_fruit_labels(subdir='train')
```

```
x_fruits_train_color=np.array(x_fruits_train_color)
y_fruits_train=np.array(y_fruits_train)

x_fruits_train_color.shape, y_fruits_train.shape
```

```
def ClassNumberToFruitName(namber):
  listaNames = ['manzana', 'platano','naranja']
  return listaNames[namber]
```
